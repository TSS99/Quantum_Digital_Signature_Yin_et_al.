## Quick Symbol Guide

- `bM`: document length in bits
- `bH`: hash length for the signature
- `bH_prime`: authentication-tag length
- `XA`, `XB`, `XC`: Alice, Bob, and Charlie key shares
- `p_a`: fresh polynomial used in one signing run
- `Sig`: the public signature Alice sends

If you get lost, focus on the flow instead of the formulas: choose parameters, sign, exchange authenticated data, verify, then check agreement.


```python
from qds_beginner import CONFIG, compute_security_params, run_qds_protocol

print("Default project configuration:")
print(CONFIG)
```

## Step 1: Choose a small document size

For a beginner demo, `bM = 500` is large enough to be realistic but still easy to inspect.


```python
document = "Hello, QDS!"
bM = 500
epsilon_target = 1e-10

params = compute_security_params(bM, epsilon_target)
params
```

## Step 2: Run the protocol

This executes the same core flow as the research notebook: simulated QKD key setup, Alice signing, authenticated exchange, and independent verification by Bob and Charlie.


```python
result = run_qds_protocol(
    document=document,
    bM=bM,
    bH=params["bH"],
    bH_prime=params["bH_prime"],
    verbose=True,
)

{key: result[key] for key in ["success", "bob_accept", "charlie_accept", "agreement", "bH", "bH_prime", "lS", "lP"]}
```

## Step 3: Try your own message

Now that one demo run succeeded, change the message or bit length and rerun the protocol. This is the fastest way to build intuition.


```python
custom_document = "Quantum signatures feel less mysterious now."
custom_bM = 768

custom_params = compute_security_params(custom_bM, epsilon_target)
custom_result = run_qds_protocol(
    document=custom_document,
    bM=custom_bM,
    bH=custom_params["bH"],
    bH_prime=custom_params["bH_prime"],
    verbose=False,
)

{key: custom_result[key] for key in ["success", "bH", "bH_prime", "lS", "lP"]}
```
